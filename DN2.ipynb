Question 1

```python
if __name__ == '__main__':
  aaup_df = pd.read_csv('aaup.data',header=None)

aaup_df[4] = pd.to_numeric(aaup_df[4],errors='coerce')
aaup_df.dropna(subset=[4],inplace=True)
ideal_sal_df = aaup_df[aaup_df[4] > 750]

y = ideal_sal_df[2].value_counts()
df = y.reset_index(name = 'Number of colleges')

df.plot(kind='bar',x='index',y='Number of colleges',figsize = (8,4),legend=False)
plt.xlabel('State(postal code)')
plt.ylabel('Number of colleges')
plt.title('Graph representing the numbers of ideal colleges from different states')
plt.show()
```

Question 2

```python
for i in [5,6,8,9]:
  aaup_df[i] = pd.to_numeric(aaup_df[i],errors='coerce')
  aaup_df.dropna(subset=[i],inplace=True)

aso_sum = aaup_df[5].sum()
aso_salr = aso_sum/len(aaup_df)

aso_sum2 = aaup_df[8].sum()
aso_comp = aso_sum2/len(aaup_df)

ast_sum = aaup_df[6].sum()
ast_salr = ast_sum/len(aaup_df)

ast_sum2 = aaup_df[9].sum()
ast_comp = ast_sum2/len(aaup_df)

aso_df = aaup_df[aaup_df[5] > aso_salr]
ast_df = aaup_df[aaup_df[6] > ast_salr]
final_aso = aso_df[aso_df[8] > aso_comp]
final_ast = ast_df[ast_df[9] > ast_comp]

print(len(final_ast)/len(ast_df))
print(len(final_aso)/len(aso_df))
```

```python
aaup_df[5] = pd.to_numeric(aaup_df[5],errors='coerce')
aaup_df[9] = pd.to_numeric(aaup_df[9],errors='coerce')
aaup_df.dropna(subset=[5,9],inplace=True)

sal_sum = aaup_df[5].sum()
avg_salr = sal_sum/len(aaup_df)

ideal_sal_df = aaup_df[aaup_df[5] > avg_salr]
ideal_clg = ideal_sal_df[ideal_sal_df[2] == "ME"]

X = np.arange(len(ideal_clg))
plt.figure(figsize = (13,5))
plt.bar(X  - 0.2, ideal_clg[5], 0.4, label = 'Average salary of associate professors')
plt.bar(X  + 0.2, ideal_clg[9], 0.4, label = 'Average compensation of associate professors')

plt.xticks(X,ideal_clg[1])
plt.xlabel("Ideal colleges in ME")
plt.ylabel("salary and compensation")
plt.title("Comparission of average salaries and average compensations of associate professors from ideal institues in ME")
plt.legend()
plt.show()
```

```python
if __name__ == '__main__':
  usnews_df = pd.read_csv('usnews.data',header=None)

for i in range(24,29):
  usnews_df[i] = pd.to_numeric(usnews_df[i],errors='coerce')

pub_df = usnews_df[usnews_df[3] == 1]
pub = []
for i in range(24,29):
  pub.append(pub_df[i].mean(axis=0,skipna=True))

pri_df = usnews_df[usnews_df[3] == 2]
pri = []
for i in range(24,29):
  pri.append(pri_df[i].mean(axis=0,skipna=True))

X = np.arange(5)
plt.figure(figsize = (13,3))
plt.bar(X  - 0.2, pub, 0.4, label = 'Average expenses of students in public colleges')
plt.bar(X  + 0.2, pri, 0.4, label = 'Average expenses of students in private colleges')

plt.xticks(X,["Room costs","Board costs","Additional fees","Estimated book costs","Estimated personal spending"])
plt.xlabel("Categories of Expenses")
plt.ylabel("Expenses in college")
plt.title("Comparision of expenses in public and private colleges")
plt.legend()
plt.show()
```
